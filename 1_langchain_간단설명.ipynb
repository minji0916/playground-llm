## [LangChain](https://python.langchain.com/v0.2/docs/introduction/)
- LangChain 은 대규모 언어 모델(LLM)을 사용하여 애플리케이션을 개발하기 위한 프레임워크
- LangServe : LangChain 체인을 REST API로 배포
- LangSmith : LLM 애플리케이션을 디버깅, 테스트, 평가, 모니터링할 수 있는 개발자 플랫폼

### LangChain 주요 기능 정리
1. 모델 & 인터페이스
2. Data Connection
3. Chains
4. Agents
5. Memory
6. Callbacks


### LangChain 주요 기능 1
모델 & 인터페이스 
- PromptTemplate, ChatPromptTemplate: 모델 입력을 템플릿화, 동적으로 선택 및 관리
- Language Models: 공통 인터페이스를 통해 언어 모델을 호출
- Output Parser: 모델 출력에서 정보 추출

### LangChain 주요 기능 2
RAG를 위한 Data Connection
- Document loaders: 다양한 소스에서 문서 불러오기
- Document transformers: 문서분할, Q&A형식으로 문서 변환, 중복 문서 작세 등 다양한 작업 수행
- Text Splitters: 문서가 너무 길어서 LLM에 한 번에 입력이 어려운 경우 사용
- Text embedding models: 비정형 텍스트를 float 숫자 목록으로 변환
- Vector stores: embedding된 데이터 저장 및 검색
- Retrievers: 문자열을 받아서 document 목록을 반환


### LangChain 주요 기능 3
- Chains : 뜻 그대로 "input -> LLM1 -> LLM2 … -> plugin -> output 등 다양한 방식으로 모듈들을 엮을 수 있습니다.
- LLM을 한 개 엮을 수도, 여러개 연결할 수도 있음
- 예시 :  chain = prompt | model 

### LangChain 주요 기능 4
Agents
- 구글 검색 결과를 활용할 수 있는 serpapi, 날씨 정보를 알 수 있는 openweather api 등 다양한 tool을 사용해서 외부 데이터를 활용할 수 있음
- 이러한 다양한 tool을 사용해서 사용자의 입력에 따라 상황에 맞게 다른 결과를 반환


### LangChain 주요 기능 5
Memory
- LLM이 데이터를 저장하고 검색 가능하도록 '기억'이 생기는 것
- 기존 채팅 기록들을 prompt에 넣어 제공하도록 구성
- 단기, 장기 기억이 가능하며 연속적인 채팅이 가능

### LangChain 주요 기능 6
Callbacks
- 모델 호출 과정에서 발생하는 이벤트를 처리하는 기능
- 예시 : 모델 호출 시 로그 출력, 중간 결과 저장, 오류 처리 등
